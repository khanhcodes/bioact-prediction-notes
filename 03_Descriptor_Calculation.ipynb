```python
from padelpy import from_smiles

def to_fp(input):
    for i in input['canonical_smiles']:
        fingerprint = from_smiles(i, fingerprints=True, descriptors=False, timeout=100)
        fingerprints = []
        x = fingerprints.append(fingerprint)
        
    return x
        
# only calculate fingerprints


# save descriptors to a CSV file
```

```python
data_fingerprints = to_fp(df3_selection)
data_fingerprints
```
